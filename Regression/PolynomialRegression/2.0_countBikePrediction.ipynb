```python
"""

Statement  : For a given dataset predict number of bikes getting shared based on different parameters 
DataSet    : https://drive.google.com/open?id=1ohN2o3zSZ2Xuy4CIdTWN-dTmciUXizst


"""

import pandas as pd
import numpy as np

bikedata=pd.read_csv('bike_sharing.csv')
print(bikedata)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

bikedata['dteday']=le.fit_transform(bikedata['dteday'])
x=bikedata.iloc[:,1:2].values
y=bikedata.iloc[:,-1].values
print(x,y)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)

import matplotlib.pyplot as plt

plt.scatter(x,y,color='Green')
plt.plot(x, reg.predict(x))
plt.title('Linear Regression')
plt.xlabel(" Date and day ")
plt.ylabel("Count of Bike ")
plt.show()
```

```python
manualPredict=reg.predict([[222]])
print(manualPredict)

from sklearn.preprocessing import PolynomialFeatures

polyReg=PolynomialFeatures(degree=2)
xPoly=polyReg.fit_transform(x)
print(xPoly)
```

```python
linreg=LinearRegression()
linreg.fit(xPoly,y)

plt.scatter(x,y,color='yellow')
plt.plot(x,linreg.predict(polyReg.fit_transform(x)))
plt.title(" Poly Regression Degree 2 ")
plt.xlabel(" Date and Day ")
plt.ylabel(" Count of Bike ")
plt.show()
```

```python
newSalary=linreg.predict(polyReg.fit_transform([[222]]))
print("Prediction of Poly Regression : ", newSalary)
```

```python
polyReg3=PolynomialFeatures(degree=3)
xPoly3=polyReg3.fit_transform(x)

linreg.fit(xPoly3,y)
plt.scatter(x,y, color="red")
plt.plot(x, linreg.predict(polyReg3.fit_transform(x)))
plt.title("Poly Regression Degree 3")
plt.xlabel("Data and Day")
plt.ylabel("Count")
plt.show()


newsalarypred3 = linreg.predict(polyReg3.fit_transform([[222]]))
print('The predicted bike Count  is ',newsalarypred3)
```

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(x)

 # Step 6 - Passing X_poly to LinearRegression
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)

# Step 7 - Visualize Poly Regression Results
plt.scatter(x,y, color="red")
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)))
plt.title("Poly Regression Degree 4")
plt.xlabel("Date and Day")
plt.ylabel("Count of bike")
plt.show()

# Step 8 Polynomial Regression prediction
new_salary_pred = lin_reg_2.predict(poly_reg.fit_transform([[522]]))
print('The predicted count of bike on date ',new_salary_pred)
```
