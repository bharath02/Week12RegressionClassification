```python
"""
Problem : Build a machine learning model to predict profit of the company  based on different expenses for a given dataset
DataSet : https://drive.google.com/open?id=1dic5dbHugytxCO6i9pcD6RPod_sLkAUA  

"""

import pandas as pd
import numpy as np

companyData=pd.read_csv("ProfitBusiness.csv")

print(companyData.shape)
print(companyData.describe())
print(companyData.isnull().sum())
```

```python
import matplotlib.pyplot as plt

companyData.plot(x='R&D Spend',y='Profit')
plt.title(" Profit of a Company on R&D Spend ")
plt.xlabel(" R&D Spend ")
plt.ylabel(" Profit ")
```

```python
companyData.plot(x='Administration',y='Profit')
plt.title(" Profit of a Company Administration")
plt.xlabel(" Administration ")
plt.ylabel(" Profit ")
```

```python
companyData.plot(x='Marketing Spend',y='Profit')
plt.title(" Profit of a Company Marketing Spend")
plt.xlabel(" Marketing Spend ")
plt.ylabel(" Profit ")
```

```python
companyData.plot(x='State',y='Profit')
plt.title(" Profit of a Company State")
plt.xlabel(" State ")
plt.ylabel(" Profit ")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
companyData['State']=le.fit_transform(companyData[['State']])


col=['R&D Spend','Administration','Marketing Spend','State']
val=['Profit']
for item in col:
    print(item, ":-")
    print(companyData[item].value_counts(),"\n")
from sklearn.model_selection import train_test_split

x=companyData[col]
y=companyData[val]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

from sklearn.linear_model import LinearRegression

reg=LinearRegression()
reg.fit(x_train,y_train)
ypred=reg.predict(x_test)

```

```python
print(y_test['Profit'][:],'\n',ypred)
```
